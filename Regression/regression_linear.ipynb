```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split
from math import sqrt

%matplotlib inline
```

```python
budget = pd.read_csv("../../../ai_algos/data/Advertising Budget and Sales.csv",index_col=0)
budget.head()
```

```python
print(budget.shape)
fig,axs = plt.subplots(1,3,sharey=True)
budget.plot(kind="scatter",x='TV Ad Budget ($)',y='Sales ($)',ax=axs[0],figsize=(16,8))
budget.plot(kind="scatter",x="Radio Ad Budget ($)",y='Sales ($)',ax=axs[1])
budget.plot(kind="scatter",x="Newspaper Ad Budget ($)",y='Sales ($)',ax=axs[2])
```

```python
#objects to train and test 
x_features = budget[["TV Ad Budget ($)","Radio Ad Budget ($)","Newspaper Ad Budget ($)"]]
y_target = budget[["Sales ($)"]]

x_train,x_test,y_train,y_test = train_test_split(x_features,y_target,random_state=1)

print(x_train.shape)
print(y_train.shape)
print(y_test.shape)
print(x_test.shape)
```

```python
linreg = LinearRegression()
linreg.fit(x_train,y_train)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
y_pred = linreg.predict(x_test)
print(y_pred)
```

```python
#Calculate the MSE
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```
