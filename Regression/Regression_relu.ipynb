```
from pandas import read_csv
from keras.models import Sequential
from keras.layers import Dense
import numpy
from matplotlib import pyplot as plt 
```

```
cd /content/drive/"My Drive"/"Colab Notebooks"
```

```
# load dataset
dataframe = read_csv("data/train_Level2Data.csv")
dataset = dataframe.values
# split into input (X) and output (Y) variables

X = dataset[:,0]
Y = dataset[:,1]
```

```
plt.title("Level2") 
plt.xlabel("x axis") 
plt.ylabel("y axis") 
plt.plot(X,Y,'ob') 
plt.show()
```

```
def CreateModel():
	# create model
	# this create the base model object
	model = Sequential()
	# add layers.
	model.add(Dense(100,input_dim=1,kernel_initializer='normal',activation='relu'))
	model.add(Dense(50,kernel_initializer='normal',activation='relu'))
	model.add(Dense(25,kernel_initializer='normal',activation='relu'))
	model.add(Dense(1,kernel_initializer='normal'))
	# Compile model
	model.compile(loss='mean_squared_error', optimizer='adam')
	return model
```

```
# use the function to create a model
model = CreateModel()

model.fit(X,Y,epochs=1000,batch_size=100,verbose=1)
```

```
# load in the test daa
dataframeTest = read_csv('data/test_level2data.csv')
datasetTest = dataframeTest.values
# get the X values
XTest = datasetTest[:,0]
```

```

YTest = model.predict(XTest)
```

```
plt.title("Level1") 
plt.xlabel("x axis") 
plt.ylabel("y axis") 
plt.plot(XTest,YTest,'ob') 
plt.show()
```

```

dataframeTest.insert(1,'Y',YTest,True)

dataframeTest.to_csv('Task3Level2.csv',sep=',',header=True,index=False)
```
